````python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
````

We use Polish mortality data, adapted from: http://stat.gov.pl/obszary-tematyczne/ludnosc/trwanie-zycia/.

````python
kobiety = pd.read_csv("data/smiertelnosc_polska_kobiety.csv", index_col="Wiek")
````

````python
kobiety.head()
````

````python
kobiety.plot(logy=False, figsize=(10,6))
````

````python
kobiety.plot(logy=True, figsize=(10,6), colormap="coolwarm")
````

````python
kobiety.transpose().plot(logy=True, figsize=(10,6), colormap="coolwarm_r")
````

````python
import numpy
````
